```python
tourney_23 = wiaa_gbb_tournament_seeding[wiaa_gbb_tournament_seeding.Year==2023].reset_index(drop=True)
tourney_23.head(10)
```

```python
# Overwrite that row with the columns for Racine St. Catherines, row 347
seeds_results_23.iloc[347, 6:] = seeds_results_23.iloc[484, 6:]
# Drop the last row
seeds_results_23 = seeds_results_23.iloc[:484, :]
```

Okay, let's try to make this functional:

```python
# Set an empty dataframe for saving data
seeding_data = pd.DataFrame({
    "Team": [],
    "Wins": [],
    "Losses": [],
    "Points_For": [],
    "Points_Against": [],
    "Win_Percentage": []
})
# Loop through all schools
for i in range(len(wiaa_gbb2223)):
    # Grab each record individually
    record = wiaa_gbb2223.iloc[i, :]
    # Set counters
    wins = 0
    losses = 0
    pts_for = 0
    pts_agn = 0
    # Set teams played
    teams_played_all = record.iloc[3].split("', ")
    # Clean the strings
    for j in range(len(teams_played_all)):
        teams_played_all[j] = str(teams_played_all[j]).replace("[", "")
        teams_played_all[j] = str(teams_played_all[j]).replace("]", "")
        teams_played_all[j] = str(teams_played_all[j]).replace("'", "")
    # See if that worked
    #print(teams_played_all)
    # Get all the scores
    game_scores = record.iloc[4].split(", ")
    # Clean scores
    for j in range(len(game_scores)):
        game_scores[j] = str(game_scores[j]).replace("[", "")
        game_scores[j] = str(game_scores[j]).replace("]", "")
        game_scores[j] = str(game_scores[j]).replace("'", "")
    # See if that worked
    #print(game_scores)
    # Set dates of games
    date_of_games = record.iloc[5].split(", ")
    # Clean dates
    for j in range(len(date_of_games)):
        date_of_games[j] = str(date_of_games[j]).replace("[", "")
        date_of_games[j] = str(date_of_games[j]).replace("]", "")
        date_of_games[j] = str(date_of_games[j]).replace("'", "")
        try:
            date_of_games[j] = datetime.strptime(str(date_of_games[j]), '%m/%d/%Y')
        except ValueError:
            date_of_games[j] = datetime.strptime("04/01/2023", '%m/%d/%Y')
    # Print the team
    team = str(record.iloc[0])
    #print(team)
    # Loop through to get wins and losses prior to the seeding date
    for j in range(len(teams_played_all)):
        # Check the date to get games considered in seeding - any games after 02/13/2023 will not be considered
        try:
            if datetime(2023, 2, 12, 0, 0) >= date_of_games[j]:
                # Split the score
                team_score = game_scores[j].split("-")[0]
                opponent_score = game_scores[j].split("-")[1]
                # Check who won:
                if team_score > opponent_score:
                    wins += 1
                elif opponent_score > team_score:
                    losses += 1
                else:
                    print("No ties in basketball")
                # Add points for and points against on season
                pts_for += float(team_score)
                pts_agn += float(opponent_score)
            else:
                continue
        except ValueError:
            print("Invalid date value")
    # Set the values of the dataframe
    seeding_data.at[i, "Team"] = str(team)
    seeding_data.at[i, "Wins"] = wins
    seeding_data.at[i, "Losses"] = losses
    seeding_data.at[i, "Points_For"] = pts_for
    seeding_data.at[i, "Points_Against"] = pts_agn
    try:
        seeding_data.at[i, "Win_Percentage"] = round((wins / (wins+losses)), 5)
    except ZeroDivisionError:
        seeding_data.at[i, "Win_Percentage"] = 0.
# See filled out dataframe
seeding_data
```

Okay, we were able to go through and figure out the winners and losers of each game. Now we need to calculate:
- Opponents' winning percentage
- Opponents' opponents' winning percentage
- Defeated opponents' winning percentage
- Defeated opponents' opponents' winning percentage

How do we calculate that?
- Go down the list of seeded teams and if the team is in the list, add it to opponents' wins and losses column 
  - If a team is not in the list, give +10 wins and +10 losses
- Defeated opponents-same idea for the first point
  - If a defeated team is not in the list, give +10 wins and +10 losses
  
Once we have calculated this, how can we the opponents' opponents' wins and losses and defeated opponents' opponents' wins and losses?
- Take the list of opponents' opponents' and add their W-L record
  - Treat the opponents' opponents' that are out of state and add 10 wins and 10 losses
- Defeated opponents' opponents' win% is the same

```python
# Recalibrate the opponents' wins and losses
seeding_data["Opponents_Wins"] = 0.
seeding_data["Opponents_Losses"] = 0.
# Loop through all schools
for i in range(len(wiaa_gbb2223)):
    # Grab each record individually
    record = wiaa_gbb2223.iloc[i, :]
    # Set counters
    wins = 0
    losses = 0
    pts_for = 0
    pts_agn = 0
    # Set teams played
    teams_played_all = record.iloc[3].split("', ")
    # Clean the strings
    for j in range(len(teams_played_all)):
        teams_played_all[j] = str(teams_played_all[j]).replace("[", "")
        teams_played_all[j] = str(teams_played_all[j]).replace("]", "")
        teams_played_all[j] = str(teams_played_all[j]).replace("'", "")
    # Get the set to get a unique list and cast it back to a list
    teams_played_unique = list(set(teams_played_all))
    # See if that worked
    #print(game_scores)
    # Set dates of games
    date_of_games = record.iloc[5].split(", ")
    # Clean dates
    for j in range(len(date_of_games)):
        date_of_games[j] = str(date_of_games[j]).replace("[", "")
        date_of_games[j] = str(date_of_games[j]).replace("]", "")
        date_of_games[j] = str(date_of_games[j]).replace("'", "")
        try:
            date_of_games[j] = datetime.strptime(str(date_of_games[j]), '%m/%d/%Y')
        except ValueError:
            date_of_games[j] = datetime.strptime("04/01/2023", '%m/%d/%Y')
    # Print the team
    team = str(record.iloc[0])
    #print(team)
    # Loop through to get wins and losses prior to the seeding date
    for j in range(len(teams_played_unique)):
        # Check the date to get games considered in seeding - any games after 02/13/2023 will not be considered
        try:
            if datetime(2023, 2, 12, 0, 0) >= date_of_games[j]:
                try:
                    # Get the opponents' record
                    opponent_wins = seeding_data[seeding_data["Team"] == str(teams_played_unique[j])]["Wins"].values[0]
                    wins += opponent_wins
                    #print(opponent_wins)
                    opponent_losses = seeding_data[seeding_data["Team"] == str(teams_played_unique[j])]["Losses"].values[0]
                    losses += opponent_losses
                # Case where team wasn't seeded or out of state
                except IndexError:
                    wins += 10
                    losses += 10
                #print(opponent_losses)
            else:
                continue
        except ValueError:
            print("Invalid date value")
    # Set the values of the dataframe
    seeding_data.at[i, "Opponents_Wins"] += wins
    seeding_data.at[i, "Opponents_Losses"] += losses
# See filled out dataframe
seeding_data
```

```python
seeding_data["Opponents_Win_Percentage"] = round(seeding_data["Opponents_Wins"] / (seeding_data["Opponents_Wins"] + seeding_data["Opponents_Losses"]), 5)
seeding_data
```

```python
# Recalibrate the opponents' wins and losses
seeding_data["Defeated_Opponents_Wins"] = 0.
seeding_data["Defeated_Opponents_Losses"] = 0.
# Loop through all schools
for i in range(len(wiaa_gbb2223)):
    # Grab each record individually
    record = wiaa_gbb2223.iloc[i, :]
    # Set counters
    wins = 0
    losses = 0
    pts_for = 0
    pts_agn = 0
    # Set teams played
    teams_played_all = record.iloc[3].split("', ")
    # Clean the strings
    for j in range(len(teams_played_all)):
        teams_played_all[j] = str(teams_played_all[j]).replace("[", "")
        teams_played_all[j] = str(teams_played_all[j]).replace("]", "")
        teams_played_all[j] = str(teams_played_all[j]).replace("'", "")
    # See if that worked
    #print(teams_played_all)
    # Get all the scores
    game_scores = record.iloc[4].split(", ")
    # Clean scores
    for j in range(len(game_scores)):
        game_scores[j] = str(game_scores[j]).replace("[", "")
        game_scores[j] = str(game_scores[j]).replace("]", "")
        game_scores[j] = str(game_scores[j]).replace("'", "")
    # See if that worked
    #print(game_scores)
    # Set dates of games
    date_of_games = record.iloc[5].split(", ")
    # Clean dates
    for j in range(len(date_of_games)):
        date_of_games[j] = str(date_of_games[j]).replace("[", "")
        date_of_games[j] = str(date_of_games[j]).replace("]", "")
        date_of_games[j] = str(date_of_games[j]).replace("'", "")
        try:
            date_of_games[j] = datetime.strptime(str(date_of_games[j]), '%m/%d/%Y')
        except ValueError:
            date_of_games[j] = datetime.strptime("04/01/2023", '%m/%d/%Y')
    # Print the team
    team = str(record.iloc[0])
    teams_defeated = []
    #print(team)
    # Loop through to get the teams defeated before the seeding date
    for j in range(len(teams_played_all)):
        # Check the date to get games considered in seeding - any games after 02/13/2023 will not be considered
        try:
            if datetime(2023, 2, 12, 0, 0) >= date_of_games[j]:
                # Split the score
                team_score = game_scores[j].split("-")[0]
                opponent_score = game_scores[j].split("-")[1]
                # Check who won:
                if team_score > opponent_score:
                    teams_defeated.append(teams_played_all[j])
            else:
                continue
        except ValueError:
            print("Invalid date or team value")
    # get the unique teams defeated
    defeated_unique = list(set(teams_defeated))
    # Loop through to get the defeated teams records
    for j in range(len(defeated_unique)):
        # Check the date to get games considered in seeding - any games after 02/13/2023 will not be considered
        try:
            if datetime(2023, 2, 12, 0, 0) >= date_of_games[j]:
                try:
                    # Get the opponents' record
                    do_wins = seeding_data[seeding_data["Team"] == str(defeated_unique[j])]["Wins"].values[0]
                    wins += do_wins
                    #print(opponent_wins)
                    do_losses = seeding_data[seeding_data["Team"] == str(defeated_unique[j])]["Losses"].values[0]
                    losses += do_losses
                # Case where team wasn't seeded or out of state, defeated teams with 10-10 record
                except IndexError:
                    wins += 10
                    losses += 10
                #print(opponent_losses)
            else:
                continue
        except ValueError:
            print("Invalid date value")
    # Set the values of the dataframe
    # Set the values of the dataframe
    seeding_data.at[i, "Defeated_Opponents_Wins"] += wins
    seeding_data.at[i, "Defeated_Opponents_Losses"] += losses
# See filled out dataframe
seeding_data
```
